```python
import pandas as pd
import os
import numpy as np
os.chdir(r"C:\Users\Bolorerdene\Downloads\bie daalt1")
irts=pd.read_excel('bie daalt.xlsx',sheet_name='Ирц')
irts=irts.dropna(axis=1, how='all')
irts['Нийт 10']=(irts[irts == 'и'].count(axis=1)*10/32)
irts
```

```python
os.chdir(r"C:\Users\Bolorerdene\Downloads\bie daalt1")
sem=pd.read_excel('bie daalt.xlsx',sheet_name='Семинар')
sem=sem.dropna(axis=1, how='all')
sem=sem.fillna(0)
sem['Нийт -25'] = (sem['Seminar1']+sem['Seminar2']+sem['Seminar3']+sem['Seminar4']+sem['Seminar5']+sem['Seminar6']+sem['Seminar7']+sem['Seminar8']+sem['Seminar9']+sem['Seminar10']+sem['Seminar11']+sem['Seminar12']+sem['Seminar13']+sem['Seminar14']+sem['Seminar15']+sem['Seminar16']+sem['Seminar17']+sem['Seminar18']+sem['Seminar19'])*25/19
sem
```

```python
os.chdir(r"C:\Users\Bolorerdene\Downloads\bie daalt1")
shalgalt=pd.read_excel('bie daalt.xlsx',sheet_name='Шалгалт')
shalgalt=shalgalt.dropna(axis=1, how='all')
shalgalt=shalgalt.fillna(0)
shalgalt['Нийт'] = shalgalt['Шалгалт - 1']+shalgalt['Шалгалт - 2']+shalgalt['Шалгалт - 3']
shalgalt
```

```python
os.chdir(r"C:\Users\Bolorerdene\Downloads\bie daalt1")
biedaalt=pd.read_excel('bie daalt.xlsx',sheet_name='Бие даалт')
biedaalt=biedaalt.dropna(axis=1, how='all')
biedaalt=biedaalt.fillna(0)
biedaalt['Нийт'] = biedaalt['Бие даалт - 1']+biedaalt['Бие даалт - 2']
biedaalt
```

```python
os.chdir(r"C:\Users\Bolorerdene\Downloads\bie daalt1")
xx=pd.read_excel('bie daalt.xlsx',sheet_name='Хүмүүжил хандлага')
xx=xx.dropna(axis=1, how='all')
xx=xx.fillna(0)
xx['Нийт'] = xx['ХХ-1']+xx['ХХ-2']+xx['ХХ-3']+xx['ХХ-4']
xx
```

```python
NO=pd.read_excel('bie daalt.xlsx',sheet_name='Нэмэлт оноо')
NO=NO.dropna(axis=1, how='all')
NO=NO.fillna(0)
NO['Нийт']=NO['Нэмэлт оноо']
NO
```

```python
niit=pd.read_excel('bie daalt.xlsx',sheet_name='Нийт')
niit=niit.dropna(axis=1, how='all')
niit=niit.dropna(axis=0,how= 'all')

niit['Ирц-10']=irts['Нийт 10']
niit['Семинар-25']=sem['Нийт -25']
niit['Шалгалт-30']=shalgalt['Нийт']
niit['Бие даалт-25']=biedaalt['Нийт']
niit['ХХ-10']=xx['Нийт']
niit['Нэмэлт оноо']=NO['Нийт']
niit['Нийт']=niit['Семинар-25']+niit['Шалгалт-30']+niit['Бие даалт-25']+niit['ХХ-10']+niit['Нэмэлт оноо']+irts['Нийт 10']

niit
```

```python
niit['Unnamed: 11']=np.where((niit['Нийт'] < 59.5 ),'F',
                           np.where((niit['Нийт'] >= 59.5 )&(niit['Нийт'] < 69.5 ),'D',
                            np.where((niit['Нийт'] >= 69.5 )&(niit['Нийт'] < 79.5 ),'C',
                            np.where((niit['Нийт'] >= 79.5 )&(niit['Нийт'] < 89.5 ),'B',
                            np.where((niit['Нийт'] >= 89.5 ),'A','NAN'
                           )))))
niit=niit.dropna(axis=1,how= 'all')
niit
```

```python
a=niit.groupby(['Unnamed: 11'])['Unnamed: 11'].count()
for i in range(len(a)):
    niit['Unnamed: 13'][i]=a[i]
niit=niit.fillna(' ')
niit
```

```python
irts_chanar = pd.Series({'Хамгийн өндөр дүн': irts['Нийт 10'].max(), 'Хамгийн бага дүн':irts['Нийт 10'].min(), 'Дундаж':irts['Нийт 10'].mean()})
sem_chanar = pd.Series({'Хамгийн өндөр дүн': sem['Нийт -25'].max(), 'Хамгийн бага дүн':sem['Нийт -25'].min(), 'Дундаж':sem['Нийт -25'].mean()})
shalg_chanar = pd.Series({'Хамгийн өндөр дүн': shalgalt['Нийт'].max(), 'Хамгийн бага дүн':shalgalt['Нийт'].min(), 'Дундаж':shalgalt['Нийт'].mean()})
biedaalt_chanar = pd.Series({'Хамгийн өндөр дүн': biedaalt['Нийт'].max(), 'Хамгийн бага дүн':biedaalt['Нийт'].min(), 'Дундаж':biedaalt['Нийт'].mean()})
xx_chanar = pd.Series({'Хамгийн өндөр дүн': xx['Нийт'].max(), 'Хамгийн бага дүн':xx['Нийт'].min(), 'Дундаж':xx['Нийт'].mean()})
NO_chanar = pd.Series({'Хамгийн өндөр дүн': NO['Нийт'].max(), 'Хамгийн бага дүн':NO['Нийт'].min(), 'Дундаж':NO['Нийт'].mean()})
niit_chanar = pd.Series({'Хамгийн өндөр дүн': niit['Нийт'].max(), 'Хамгийн бага дүн':niit['Нийт'].min(), 'Дундаж':niit['Нийт'].mean()})
newSheetScore = pd.DataFrame({'Ирц':irts_chanar,'Семинар': sem_chanar, 'Шалгалт':shalg_chanar,'Бие Даалт': biedaalt_chanar, 'Хүмүүжил Хандлага':xx_chanar, 'Нэмэлт Оноо': NO_chanar, 'Нийт дүн':niit_chanar})
newSheetScore
```

```python
s=niit['Unnamed: 13'][4]/niit['Unnamed: 13'][5]
d=(niit['Unnamed: 13'][0]+niit['Unnamed: 13'][1])/niit['Unnamed: 13'][5]
hich_chanar= pd.Series({'Хичээлийн амжилт': s , 'Хичээлийн чанар':d })

bolor=pd.concat([hich_chanar,newSheetScore])
bolor=bolor.fillna(' ')
bolor
```

```python
writer = pd.ExcelWriter('biedaalt_b20fa1526.xlsx', engine = 'xlsxwriter')
irts.to_excel(writer, sheet_name='Ирц')
sem.to_excel(writer, sheet_name='Семинар')
shalgalt.to_excel(writer, sheet_name='Шалгалт')
biedaalt.to_excel(writer, sheet_name='Бие Даалт')
xx.to_excel(writer, sheet_name='Хүмүүжил Хандлага')
NO.to_excel(writer, sheet_name='Нэмэлт оноо')
niit.to_excel(writer, sheet_name='Нийт')
bolor.to_excel(writer, sheet_name='Нэмэлт тооцоолол')
writer.save()
print('Successful')
```
